# EDA Visualisasi

```python
# Sales dan profit per region
sales_profit_region = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
```

```python
# Sales dan Profit per kategori
sales_profit_category = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
```

```python
# Korelasi antara sales, discount, dan profit
correlation = df[['Sales', 'Discount', 'Profit']].corr()
```

```python
# Visualisasi Outlier
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Category', y='Profit', palette='Set3')
plt.title('Boxplot Outlier Profit per Kategori')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Category', y='Sales', hue='Category', palette='Set2', legend=False)
plt.title('Boxplot Outlier Sales per Kategori')
plt.tight_layout()
plt.show()
```

```python
# Distribusi kategori produk
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Category', palette='pastel')
plt.title('Distribusi Kategori Produk')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Frekuensi jumlah transaksi dalam setiap produk tertinggi adalah kategori office supplies

```python
# Sales vs Profit per Region
plt.figure(figsize=(8, 5))
sns.barplot(data=sales_profit_region, x='Region', y='Sales', color='skyblue', label='Sales')
sns.barplot(data=sales_profit_region, x='Region', y='Profit', color='orange', label='Profit')
plt.title('Total Sales dan Profit per Region')
plt.legend()
plt.tight_layout()
plt.show()
```

Wilayah dengan total penjualan dan pendapatan di dominasi oleh East dan West

```python
# Sales vs Profit per Kategori
plt.figure(figsize=(6, 4))
sns.barplot(data=sales_profit_category, x='Category', y='Sales', color='lightblue', label='Sales')
sns.barplot(data=sales_profit_category, x='Category', y='Profit', color='coral', label='Profit')
plt.title('Sales dan Profit per Kategori Produk')
plt.legend()
plt.tight_layout()
plt.show()
```

Technology dan office supplies memperoleh total penjualan dan total pendapatan lebih banyak dibandingkan dengan furniture

```python
# Trend Penjualan per Bulan
monthly_sales = df.set_index('Order Date').resample('M')['Sales'].sum()
plt.figure(figsize=(10, 5))
monthly_sales.plot(marker='o', color='purple')
plt.title('Trend Penjualan per Bulan')
plt.ylabel('Sales')
plt.xlabel('Bulan')
plt.tight_layout()
plt.show()
```

Tren penjualan diambil berdasarkan tahun dan bulan, menunjukkan grafik yang kurang stabil di setiap tahun


```python
#Korelasi sales vs profit
plt.figure(figsize=(6, 4))
sns.lmplot(data=df, x='Sales', y='Profit', hue='Category', aspect=1.5, height=5)
plt.title('Sales vs Profit ')
plt.tight_layout()
plt.show()
```

- Terlihat hubungan positif antara Sales dan Profit, artinya semakin tinggi   penjualan, umumnya semakin tinggi keuntungan

- Kategori Technology cenderung memberikan profit lebih tinggi dibandingkan Furniture dan Office Supplies.

```python
 # Heatmap korelasi
plt.figure(figsize=(5, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi antara Sales, Discount, dan Profit')
plt.tight_layout()
plt.show()
```
